## Week 2 Lab 

Topics: Missing value handling, encoding categorical variables, scaling. 
Problems: (a) Fill missing values with mean/median/mode. (b) Apply label 
encoding and one-hot encoding. (c) Normalize/standardize numerical features.

```python
import pandas as pd
import numpy as np

# 1. Load the original dataset
df = pd.read_csv('Iris.csv')

# =====================================================================
# (a) Fill Missing Values Manually
# =====================================================================

numerical_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for col in numerical_cols:
    mean_val = df[col].mean()
    df[col] = df[col].fillna(mean_val)

# Categorical column: Fill with Mode
mode_val = df['species'].mode()[0]
df['species'] = df['species'].fillna(mode_val)
```

## Label Encoding 

```python
# =====================================================================
# (b) Manual Label Encoding 
# =====================================================================
def label_encoding(df, column_name):
    unique_items = sorted(df[column_name].unique())
    mapping = {}
    for index, item in enumerate(unique_items):
        mapping[item] = index
    print(mapping)
    df[column_name + '_labeled'] = df[column_name].map(mapping)
    return df
    
df = label_encoding(df, 'species')
```

## One hot encoding 

```python
# =====================================================================
# (b) Manual One Hot Encoding
# =====================================================================
def one_hot_encoding(df, column_name):
    unique_items = df[column_name].unique()   #---> 3 Categories 
    
    # Create a new column for each category
    for item in unique_items:
        new_col_name = f"{column_name}_{item}"
        # If the row matches the item, put 1, else 0
        df[new_col_name] = (df[column_name] == item).astype(int)   #0-->large,1-->medium,2-->small
        
    return df

df = one_hot_encoding(df, 'species')
```

## Normalization

```python
# =====================================================================
# (c) Manual Normalization
# =====================================================================
def normalization(df, columns):
    for col in columns:
        col_min = df[col].min()
        col_max = df[col].max()
        
        # Formula: (value - min) / (max - min)
        df[col + '_norm'] = (df[col] - col_min) / (col_max - col_min)
        return df

df = normalization(df, numerical_cols)
```

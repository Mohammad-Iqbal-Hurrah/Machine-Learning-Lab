## B. Load sample dataset Iris or Titanic and explore data types, nulls, and summary statistics.  

```python
# 1. Load the Iris dataset
df = pd.read_csv('Iris_Cleaned.csv')

# 2. Display the first 5 rows
print("--- FIRST 5 ROWS ---")
print(df.head())

# Check data types and null values
print(f"\nCheck Data Types and Null values")
print("\n--- DATASET INFO ---")
df.info()

# Display summary statistics for numerical columns
print(f"\nSummary Statictics For Numerical Values")
print("\n--- SUMMARY STATISTICS ---")
print(df.describe())

# Check for missing values explicitly
print(f"Missing Values per Column")
print("\n--- MISSING VALUES PER COLUMN ---")
print(df.isnull().sum())
```

## C. Visualize basic distributions using histograms and boxplots. 

```python
# Simple Histogram (Distribution of a single variable)
import matplotlib.pyplot as plt 
plt.figure(figsize=(6, 4))

plt.hist(df['sepal_length'], bins=15, color='skyblue', edgecolor='black')

plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Group your data by the category column
categories = df['species'].unique()  # Get unique species names
data_groups = [df[df['species'] == cat]['petal_width'] for cat in categories]

plt.figure(figsize=(6, 4))

# Plot and label each group
plt.boxplot(data_groups, labels=categories)

plt.title('Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width')

plt.tight_layout()
plt.show()
```
